```python
import torch
```

```python
from torch.autograd import Variable
import cv2
from data import BaseTransform, VOC_CLASSES as labelmap
from ssd import build_ssd
import imageio
```

```python
def detect(frame, net, transform):
    height, width = frame.shape[:2]
    frame_t = transform(frame)[0]
    x = torch.from_numpy(frame_t).permute(2,0,1)
    x = x.unsqueeze(0)
    x = Variable(x)
    
    y = net(x)
    
    detections = y.data
    
    scale = torch.Tensor([width, height, width, height])
    print(detections)
    for i in range(detections.size(1)):
        j=0
        while detections[0,i,j,0]>=0.5:
            pt = (detections[0,i,j,1:]*scale).numpy()
            cv2.rectangle(frame, (int(pt[0]), int(pt[1])), (int(pt[2]), int(pt[3])), (0,0,255), 2)
            cv2.putText(frame, labelmap[i-1], (int(pt[0]), int(pt[1])), cv2.FONT_HERSHEY_SIMPLEX, 2, (255,0,0), 2, cv2.LINE_AA)
            
            j += 1
    return frame

net = build_ssd('test')
net.load_state_dict(torch.load('ssd300_mAP_77.43_v2.pth', map_location= lambda storage, loc: storage))

transform = BaseTransform(net.size, (104/256.0, 117/256.0, 123/256.0))

# reader = imageio.get_reader('epic_horses.mp4')
# FPS = reader.get_meta_data()['fps']
# writer = imageio.get_writer('output1.mp4', fps = FPS)

# for i, frame in enumerate(reader):
#     frame = detect(frame, net.eval(), transform)
#     writer.append_data(frame)
#     print(i)
# writer.close


video_capture = cv2.VideoCapture(0)
while True:
    _, frame = video_capture.read()
#     gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    canvas = detect(frame, net.eval(), transform)
    cv2.imshow('Video', canvas)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

video_capture.release()
cv2.destroyAllWindows()
```
